```python
import os
import math
import re
from collections import defaultdict
```

```python


# ------------------------------
# MySet (simple wrapper over set)
# ------------------------------
class MySet:
    def __init__(self):
        self.data = set()

    def add(self, item):
        self.data.add(item)

    def union(self, other):
        return self.data.union(other.data)

    def intersection(self, other):
        return self.data.intersection(other.data)

    def get_all(self):
        return list(self.data)


# ------------------------------
# Position class
# ------------------------------
class Position:
    def __init__(self, page_name, index):
        self.page_name = page_name
        self.index = index


# ------------------------------
# WordEntry class
# ------------------------------
class WordEntry:
    def __init__(self, word):
        self.word = word
        self.positions = []  # list of Position

    def add_position(self, position):
        self.positions.append(position)

    def get_positions(self):
        return self.positions


# ------------------------------
# PageIndex class
# ------------------------------
class PageIndex:
    def __init__(self):
        self.word_entries = {}  # word -> WordEntry

    def add_position_for_word(self, word, position):
        if word not in self.word_entries:
            self.word_entries[word] = WordEntry(word)
        self.word_entries[word].add_position(position)

    def get_word_entries(self):
        return self.word_entries


# ------------------------------
# PageEntry class
# ------------------------------
class PageEntry:
    def __init__(self, page_name, file_path):
        self.page_name = page_name
        self.page_index = PageIndex()
        self.word_count = 0

        self.process_page(file_path)

    def process_page(self, file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read().lower()

        words = re.findall(r'\w+', text)

        for idx, word in enumerate(words):
            pos = Position(self.page_name, idx)
            self.page_index.add_position_for_word(word, pos)

        self.word_count = len(words)

    def get_page_index(self):
        return self.page_index


# ------------------------------
# MyHashTable (word -> WordEntry)
# ------------------------------
class MyHashTable:
    def __init__(self):
        self.table = {}

    def add_positions_for_word(self, word, word_entry):
        if word not in self.table:
            self.table[word] = WordEntry(word)

        self.table[word].positions.extend(word_entry.get_positions())

    def get_word_entry(self, word):
        return self.table.get(word, None)


# ------------------------------
# InvertedPageIndex
# ------------------------------
class InvertedPageIndex:
    def __init__(self):
        self.hash_table = MyHashTable()
        self.pages = []

    def add_page(self, page_entry):
        self.pages.append(page_entry)

        for word, word_entry in page_entry.get_page_index().get_word_entries().items():
            self.hash_table.add_positions_for_word(word, word_entry)

    def get_pages_containing_word(self, word):
        entry = self.hash_table.get_word_entry(word)
        result = set()

        if entry:
            for pos in entry.get_positions():
                result.add(pos.page_name)

        return result

    def total_pages(self):
        return len(self.pages)


# ------------------------------
# Search Engine
# ------------------------------
class SearchEngine:
    def __init__(self, dataset_path):
        self.index = InvertedPageIndex()
        self.dataset_path = dataset_path

    # load all webpages
    def load_pages(self):
        webpages_path = os.path.join(self.dataset_path, "webpages")

        for root, dirs, files in os.walk(webpages_path):
            for file in files:
                full_path = os.path.join(root, file)
                page_entry = PageEntry(file, full_path)
                self.index.add_page(page_entry)

    # TF calculation
    def tf(self, word, page):
        word_entries = page.get_page_index().get_word_entries()
        if word not in word_entries:
            return 0
        return len(word_entries[word].get_positions()) / page.word_count

    # IDF calculation
    def idf(self, word):
        df = len(self.index.get_pages_containing_word(word))
        if df == 0:
            return 0
        return math.log(self.index.total_pages() / df)

    # TF-IDF score
    def tf_idf(self, word, page):
        return self.tf(word, page) * self.idf(word)

    # phrase matching using positions
    def phrase_score(self, words, page):
        word_entries = page.get_page_index().get_word_entries()

        if any(w not in word_entries for w in words):
            return 0

        positions_lists = [word_entries[w].get_positions() for w in words]

        count = 0
        first_positions = positions_lists[0]

        for pos in first_positions:
            match = True
            current_index = pos.index

            for i in range(1, len(words)):
                next_positions = positions_lists[i]
                if not any(p.index == current_index + i for p in next_positions):
                    match = False
                    break

            if match:
                count += 1

        return count

    # search query
    def search(self, query):
        words = query.lower().split()
        scores = {}

        for page in self.index.pages:
            score = 0

            # TF-IDF scoring
            for word in words:
                score += self.tf_idf(word, page)

            # phrase boost
            score += self.phrase_score(words, page) * 2

            if score > 0:
                scores[page.page_name] = score

        # sort pages by score
        ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)

        return [page for page, _ in ranked]


# ------------------------------
# Action Handler
# ------------------------------
def process_actions(engine, actions_file):
    with open(actions_file, 'r') as f:
        lines = f.readlines()

    for line in lines:
        line = line.strip()

        if line.startswith("query"):
            query = line.replace("query", "").strip()
            result = engine.search(query)

            if result:
                print(", ".join(result))
            else:
                print("No results found")


```

```python
# ------------------------------
# MAIN
# ------------------------------
if __name__ == "__main__":
    dataset_path = "dataset/Q2_websearch"
    actions_file = os.path.join(dataset_path, "actions.txt")

    engine = SearchEngine(dataset_path)
    engine.load_pages()

    process_actions(engine, actions_file)
```
