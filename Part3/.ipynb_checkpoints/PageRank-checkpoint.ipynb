```python
from pyspark import SparkContext

sc = SparkContext.getOrCreate()
sc.setLogLevel("ERROR")
```

```python
# PageRank parameters
beta = 0.8
iterations = 40

# Small graph
file_path = "/home/rajesh/CSL7100-Assignment4/dataset/Q3_pagerank/small.txt"
n = 53   # number of nodes in small graph
```

```python
# Load file
lines = sc.textFile(file_path)

# Parse edges (source, destination)
edges = lines.map(lambda x: tuple(map(int, x.split())))

# Remove duplicate edges
edges = edges.distinct()

# Check few edges
edges.take(5)
```
