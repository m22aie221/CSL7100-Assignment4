```python
from pyspark import SparkConf, SparkContext

conf = SparkConf() \
    .setAppName("PageRank") \
    .setMaster("local[8]") \
    .set("spark.executor.memory", "4g") \
    .set("spark.driver.memory", "4g")

sc = SparkContext.getOrCreate(conf)
sc.setLogLevel("ERROR")
```

```python
from pyspark import SparkContext  
from pyspark import StorageLevel  
from operator import add  
```

```python
def load_edges(sc, file_path, num_partitions=8):
    """
    Reads file from HDFS and converts to (src, dst) edges
    """
    lines = sc.textFile(file_path, minPartitions=num_partitions)
    
    edges = (
        lines
        .map(lambda x: x.strip().split())
        .filter(lambda x: len(x) == 2)  # avoid bad lines
        .map(lambda x: (x[0], x[1]))
    )
    
    return edges
```

```python
# PageRank parameters
NUM_ITER = 40
DAMPING = 0.85

# Parallelism (based on your machine: 8 cores)
NUM_PARTITIONS = 8
```

```python
def build_graph(edges):
    """
    Creates adjacency list: node -> list of neighbors
    """
    links = edges.groupByKey().mapValues(list)
    
    # Cache because used repeatedly in iterations
    links = links.persist(StorageLevel.MEMORY_ONLY)
    
    return links
```

```python
def compute_contributions(neighbors, rank):
    """
    Distribute rank equally to neighbors
    """
    num_neighbors = len(neighbors)
    for n in neighbors:
        yield (n, rank / num_neighbors)
```

```python
def run_pagerank_convergence(links, max_iter=50, damping=0.85, tol=1e-6):
    """
    PageRank with convergence check instead of fixed iterations

    Parameters:
    - links: adjacency list RDD (node -> neighbors)
    - max_iter: maximum iterations allowed
    - damping: damping factor (usually 0.85)
    - tol: convergence threshold (smaller = stricter)
    """

    # Total number of nodes in graph
    n = links.count()

    # Initialize ranks with uniform probability (sum = 1)
    ranks = links.mapValues(lambda _: 1.0 / n)

    # Store previous iteration ranks for comparison
    prev_ranks = None

    for i in range(max_iter):

        # Step 1: Join links with current ranks
        # (node -> (neighbors, current_rank))
        joined = links.join(ranks)

        # Step 2: Distribute rank contributions to neighbors
        contributions = joined.flatMap(
            lambda x: compute_contributions(x[1][0], x[1][1])
        )

        # Step 3: Aggregate contributions and apply PageRank formula
        new_ranks = contributions.reduceByKey(add).mapValues(
            lambda rank: (1 - damping) / n + damping * rank
        )

        # Step 4: Check convergence (if not first iteration)
        if prev_ranks is not None:

            # Join current and previous ranks
            diff = new_ranks.join(prev_ranks) \
                .map(lambda x: abs(x[1][0] - x[1][1])) \
                .sum()

            print(f"Iteration {i}: diff = {diff}")

            # If change is very small → stop early
            if diff < tol:
                print(f"✅ Converged at iteration {i}")
                return new_ranks

        # Update previous ranks for next iteration
        prev_ranks = new_ranks

        # Update ranks
        ranks = new_ranks

    # If max_iter reached without convergence
    print("⚠️ Reached max iterations without full convergence")

    return ranks
```

```python
def show_results(ranks, top_k=5):
    """
    Show top and bottom ranked nodes
    """
    print("Top nodes:")
    for node, rank in ranks.takeOrdered(top_k, key=lambda x: -x[1]):
        print(node, rank)

    print("\nBottom nodes:")
    for node, rank in ranks.takeOrdered(top_k, key=lambda x: x[1]):
        print(node, rank)
```

```python
def plot_rank_distribution(ranks):
    """
    Plot histogram of PageRank values
    """
    import matplotlib.pyplot as plt
    
    # Collect ranks to driver (safe for this dataset size)
    rank_values = ranks.values().collect()
    
    plt.figure()
    plt.hist(rank_values, bins=30)
    
    plt.title("PageRank Distribution")
    plt.xlabel("Rank Value")
    plt.ylabel("Frequency")
    
    plt.show()
```

```python
FILE_PATH = "hdfs://localhost:9000/user/rajesh/dataset/Q3_pagerank/small.txt"

edges = load_edges(sc, FILE_PATH, NUM_PARTITIONS)
links = build_graph(edges)

ranks = run_pagerank_convergence(links, max_iter=40, tol = 0)

show_results(ranks)


print("Total rank:", ranks.values().sum())
```

```python
plot_rank_distribution(ranks)
```

```python
FILE_PATH = "hdfs://localhost:9000/user/rajesh/dataset/Q3_pagerank/whole.txt"

edges = load_edges(sc, FILE_PATH, NUM_PARTITIONS)
links = build_graph(edges)

ranks = run_pagerank_convergence(links, max_iter=40, tol=0)

show_results(ranks)

print("Total rank:", ranks.values().sum())
```

```python
plot_rank_distribution(ranks)
```
